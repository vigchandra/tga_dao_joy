```python
import pandas as pd
import numpy as np
```

```python
df = pd.ExcelFile('data_input/TGA Data-JT-DAO_data.xls')
```

```python
d = {}
for i in range(31):
    k = df.sheet_names[i]
    d[k] = pd.DataFrame(df.parse(i))
    d[k]['sheet_name'] = k
    d[k]['compound'] = df.sheet_names[i].split('-')[0]
    d[k]['condition'] = df.sheet_names[i].split('-')[1]
    d[k]['exp_no'] = df.sheet_names[i].split('-')[2][0]
    d[k] = d[k].drop(columns = ['Index', 'Unnamed: 0'])
    #d[k] = d[k].dropna()
    d[k] = d[k][1:]
    print("Imported ", df.sheet_names[i], d[k].shape)

df_prod = d[list(d.keys())[0]]

for i in range(1, 31):
    df_prod = df_prod.append(d[list(d.keys())[i]])
    print(df_prod.shape)
df_prod = df_prod.dropna()
print(df_prod.shape)
```

```python
df_prod.to_csv('data_input/prod_v1.csv')
```

```python
e = {}
for i in range(31, len(df.sheet_names)):
    s = df.sheet_names[i]
    e[s] = pd.DataFrame(df.parse(i))
    e[s]['sheet_name'] = s
    e[s]['compound'] = df.sheet_names[i].split('(')[0].strip()
    e[s]['compound'] = e[s]['compound'].astype('str')
    e[s]['exp_no'] = df.sheet_names[i].split('(')[1].strip()[0]
    e[s] = e[s].drop(columns = ['Index', 'Unnamed: 0'])
    e[s] = e[s][1:]
    print("Imported ", df.sheet_names[i], e[s].shape)

df_feed = e[list(e.keys())[0]]
df_feed['sheet_name'] = df.sheet_names[0]
for i in range(1,9):
    df_feed = df_feed.append(e[list(e.keys())[i]])
    print(df_feed.shape)
df_feed = df_feed.dropna()
print(df_feed.shape)
```
